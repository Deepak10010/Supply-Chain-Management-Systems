```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python

# Load the dataset
df = pd.read_csv('SCMS Dataset.csv')
df.head()
```

```python
df.info()
```

```python
df = df.drop('ID', axis=1)
```

```python
df
```

```python
top_project_codes = df['Project Code'].value_counts().index[:10]
df_top_project_codes = df[df['Project Code'].isin(top_project_codes)]
average_pack_price = df_top_project_codes.groupby('Project Code')['Pack Price'].mean()

colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'orange', 'purple', 'pink']

plt.figure(figsize=(10,5))
plt.bar(average_pack_price.index, average_pack_price.values, color=colors)
plt.xlabel('Project Code')
plt.ylabel('Average Pack Price')
plt.title('Average Pack Price by Project Code')
plt.xticks(rotation=90)
plt.show()
```

This bar chart shows the average pack price for each of the top 10 project codes. The height of each bar represents the average pack price for the corresponding project code.

From this chart, we can observe that:

The project code '103-ZW-T01' has the highest average pack price among the top 10 project codes.
The project codes '102-ZM-T01' and '103-ZW-T01' have relatively higher average pack prices compared to the other project codes.
The project code '102-ZM-T01' has the lowest average pack price among the top 10 project codes.
This information can be useful for understanding how pack prices vary across different project codes. For example, if you are interested in cost optimization, you might want to investigate why some project codes have higher average pack prices than others.

```python
# Select top 10 PQ# by number of shipments for clearer visualization
top_pq = df['PQ #'].value_counts().index[:10]
df_top_pq = df[df['PQ #'].isin(top_pq)]

average_pack_price_pq = df_top_pq.groupby('PQ #')['Pack Price'].mean()

plt.figure(figsize=(10,5))
plt.bar(average_pack_price_pq.index, average_pack_price_pq.values, color=colors)
plt.xlabel('PQ #')
plt.ylabel('Average Pack Price')
plt.title('Average Pack Price by PQ #')
plt.xticks(rotation=90)
plt.show()
```

This visualization allows us to compare the average pack price across different PQ#. As you can see, the average pack price varies significantly between different PQ#.
